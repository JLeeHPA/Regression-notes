```sas
data perch;
    set sashelp.fish;
    where species = "Perch";
    
    height2 = length3 * height/100;
    width2 = length3 * width/100;
    
    drop height width;
    rename height2 = height width2 = width;
run;
```

```sas
proc reg data = perch plots = none outest = ridge_data ridge = 40;
    model weight = length1 length2 length3 height width;
run;

proc print data = ridge_data;
run;
```
